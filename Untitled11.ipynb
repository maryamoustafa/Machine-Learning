```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from sklearn.preprocessing import StandardScaler
```

```python
def GUC_Distance(Cluster_Centroids, Data_points, Distance_Type):
    if Distance_Type.lower() == 'euclidean':
        diff = np.expand_dims(Data_points, axis=1) - np.expand_dims(Cluster_Centroids, axis=0)
        distances = np.sqrt(np.sum(diff ** 2, axis=2))
    elif Distance_Type.lower() == 'pearson':
        dp_norm = Data_points - np.mean(Data_points, axis=0)
        cc_norm = Cluster_Centroids - np.mean(Cluster_Centroids, axis=0)
        correlation = np.dot(dp_norm, cc_norm.T) / (len(Data_points) - 1)
        distances = 1 - correlation
    else:
        raise ValueError("Unsupported distance type. Choose 'euclidean' or 'pearson'.")
    return distances
```

```python
def GUC_Kmean(Data_points, Number_of_Clusters, Distance_Type, max_iterations=100, tolerance=1e-4):
    mins = Data_points.min(axis=0)
    maxs = Data_points.max(axis=0)
    rng = np.random.default_rng()
    Cluster_Centroids = mins + rng.random((Number_of_Clusters, Data_points.shape[1])) * (maxs - mins)

    for iteration in range(max_iterations):
        Cluster_Distances = GUC_Distance(Cluster_Centroids, Data_points, Distance_Type)
        closest_centroids = np.argmin(Cluster_Distances, axis=1)
        new_centroids = np.array([Data_points[closest_centroids == i].mean(axis=0) for i in range(Number_of_Clusters)])

        if np.allclose(Cluster_Centroids, new_centroids, atol=tolerance):
            break
        Cluster_Centroids = new_centroids

    Cluster_Metric = np.sum(np.min(Cluster_Distances, axis=1) ** 2)
    return closest_centroids, Cluster_Centroids, Cluster_Metric
```

```python
file_path = ('C:\\Users\\marya\\Downloads\\GUC_2350_56_17328_2022-02-22T14_59_49\\Customer data.csv') 
customer_data = pd.read_csv(file_path)

features = customer_data[['Sex', 'Marital status', 'Age', 'Education', 'Income', 'Occupation', 'Settlement size']]

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
num_clusters = 7 
labels, centroids, metric = GUC_Kmean(scaled_features, num_clusters, 'euclidean')

# Visualize Clustering Results
plt.figure(figsize=(10, 6))
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100)
plt.title("Customer Clusters")
plt.xlabel("Scaled Age")
plt.ylabel("Scaled Income")
plt.show()
```
